```python
# Librerias necesarias
import pandas as pd
from sqlalchemy import create_engine, Table, MetaData, Integer, VARCHAR, String, insert, Column, select
```

```python
# Cargamos los datos del CSV
df = pd.read_csv('../../Datasets/reportes_visible_2023-10-oct.csv', encoding='latin1')
```

```python
df.head()
```

```python
# Creamos la conexion al server de PostgreSQL
engine = create_engine("postgresql+psycopg2://postgres:123456@192.168.100.47/alerta_272")
connection = engine.connect()
metadata_obj = MetaData()
```

# Tipo de tipos_reportantes

```python
# Valores unicos
data = df.reportante.unique()
data
```

```python
tipos_reportantes = Table(
    'tipos_reportantes',
    metadata_obj,
    Column('id', Integer, primary_key = True),
    Column("tipo_reportante", VARCHAR(30)))
```

```python
# Insertando en la tabla tipos_reportantes

ins = tipos_reportantes.insert().values([{'tipo_reportante': item} for item in data])
# result = connection.execute(ins)
# print(result )
with engine.begin() as conn: 
    conn.execute(ins) 
```

# Subtipos de agresión

```python
# Valores unicos
data = df[df.tipo_de_agresion.notna() & df.subtipo_de_agresion.notna()][['tipo_de_agresion', 'subtipo_de_agresion']] # .sort_values().unique().tolist()
# data.append('Otro')
# data.sort()
data
```

```python
g = df[df.tipo_de_agresion.notna() & df.subtipo_de_agresion.notna()].groupby('tipo_de_agresion', sort=True)['subtipo_de_agresion'].agg(['unique'])
g
```

```python
for idx, item_cat in enumerate(g.index):
    items = [x for x in g.loc[item_cat].values.tolist()]
    # print(items)
    for item in items:
        for x in item:
            print("""INSERT INTO subtipos_agresion (tipo_agresion_id, subtipo_agresion) VALUES(
                (SELECT id FROM tipos_agresion WHERE tipo_agresion = '{}'), '{}'
            );""".format(item_cat, x))
        #print(f"{idx} - {item}")
```
